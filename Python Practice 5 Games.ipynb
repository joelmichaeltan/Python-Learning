### Dice Rolling Simulator

Just dices and numbers for you.

```python
#Import
from random import randint

#define a function for 1 dice roll
def roll_dice():
    roll=randint(1,6)
    return roll

#setting a counter as True
repeat = True

#Loop in order to prompt attempt again
while repeat:
    #input no. of dices
    raw=int(input("How many dices do you want? :"))
    
    #setting a list to place all the answers
    dice_list=[]

    #Iterate through with the function 1 dice roll to get multiple answers and append them to the list 
    for b in range(0,raw):
        dice_list.append(roll_dice())

    #print list of answers and print a prompt
    print(dice_list)
    print("Again? (Y/N)")
    
    #repeat while Y is True
    repeat = "Y" in input()
```

### Guess the Numbeer

Guess a number from 1 to 10. In Singapore we call this game Zhong Ji Mi Ma. 
P.S. Please do not use the word sauce lightly.


```python

#Import
from random import randint

#define a function that returns a random number from 1 to 10
def callrandnum():
    randnum=randint(1,10)
    return randnum

#Calling the function
randomnumber=callrandnum()

#Looping in order to prompt attempt again
repeat= True
while repeat:
    
    #Input Number
    raw_input=input("Guess the Number :")
    
    #Using if/elif/else to produce different result
    if raw_input == "GTFO": #I set this to get out of(stop) the game anytime
        print("okay")
        break
        
    elif raw_input.isalpha(): #If you input a string instead of a number, it will call you to do it again.
        print("Did you even put a number?!")
        repeat
        
    #If the input does not match up to the top two, then,    
    else:
        if int(raw_input)> randomnumber: #if more than random number
            print("You are too high bro\nTry again")
            repeat
        elif int(raw_input)< randomnumber: #if less than random number
            print("You are too low brah\nTry again")
            repeat
        elif int(raw_input)== randomnumber: #if guessed correctly
            print("Congratz.. You are some lucky sauce")
            break
        
    
    
    
```

```python

#Input going in
going_in=input("You are now entering the HOUSE OF THE LIVING DEAD. Are you SURE you want to go in? : (y/n)")

#if y is input
if going_in=='y':
    #setting counter for invalid_turining (just see this as something to keep the while loop going)
    invalid_turning= True 
    #looping to prompt attempt again
    while invalid_turning:
        #input turning left wing of house or right wing of house
        turning = input("Turn left or right :")
        #setting another counter for invalid_leftright (just see this as something to keep the while loop going)
        invalid_leftright= True
        #looping to prompt attempt again
        while invalid_leftright:
            #if the input is left
            if turning == 'left':
                #prompt a door choice
                leftside = input("Chose door one two or three or turn back :")
                #if door choice is one
                if leftside == 'one':
                    print("The room you entered is full of japanese hello kitty. A cute japanese man all dressed in a sexy kitty costume awaits you")
                    invalid_leftright= False
                    invalid_turning= False
                #if door choice is two
                elif leftside == 'two':
                    print("You find a big 2")
                #if door choice is three
                elif leftside == 'three':
                    print("You find a big 3")
                #if you want to go back to the middle
                elif leftside == 'turn back':
                    invalid_leftright= False
        

            #turning right
            elif turning =='right':
                
                #input choice
                rightside = input("Chose one two or three or turn back :")
                #if door choice is one
                if rightside == 'one':
                    print("you find a big upyours")     
                #if door choice is two
                elif rightside == 'two':
                    print("you find a big middle finger")
                #if door choice is three
                elif rightside == 'three':
                    print("GTFO")
                #if you want to go back to the middle
                elif rightside == 'turn back':
                    invalid_leftright= False
                
#getting out if input is not 'y' from the start        
else:
    print('goodbye')
```
